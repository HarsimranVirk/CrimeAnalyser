```python
import pandas as pd
```

```python
df = pd.read_csv('SGD Project - Sheet1.csv')
df.head()
df.shape
```

```python
df_removed_2014.isna()
```

```python
df_row_rem = df_col_rem.dropna(how='all')
df_row_rem.shape
```
