```python
import pandas as pd
```

```python
df = pd.read_csv('CleanedSheet.csv')
df.head()
```

```python
df.drop(['TOTAL IPC CRIMES'], axis=1, inplace=True)
df.head()
```

```python
def location_name(dist, state):
    return "{dist}, {state}".format(dist=dist, state=state)
```

```python
dist = df['DISTRICT']
state = df['STATE/UT']
size = len(state)
```

```python
loc = []
loc_id = []
for i in range(size):
    name = location_name(dist[i], state[i])
    loc.append(name)
    loc_id.append(i+1)
```

```python
assert len(loc) == df.shape[0]
assert "CHHINDWARA, MADHYA PRADESH" in loc
assert "LUCKNOW, Uttar Pradesh" in loc
assert "SURENDRANAGAR, GUJARAT" in loc
```

```python
df['id'] = loc_id
df['location'] = loc
df.head()
```

```python
df.to_csv("CleanedSheet.csv", index=False)
```
